```python
import csv
from datetime import datetime
```

```python
#Pizza üst sınıfını oluşturuyoruz
class Pizza:
  def __init__(self,description,price):
    self.description=description
    self.price=price

  def get_description(self):
        return self.description

  def get_cost(self):
        return self.price

#Pizza alt sınıfları

class KlasikPizza(Pizza):
    def __init__(self):
        super().__init__("Klasik Pizza", 10)


class MargaritaPizza(Pizza):
    def __init__(self):
        super().__init__("Margarita Pizza ", 11)


class TurkPizza(Pizza):
    def __init__(self):
        super().__init__("Turk Pizza ", 15)


class DominosPizza(Pizza):
    def __init__(self):
        super().__init__("Dominos Pizza ", 10)
```

```python
#Pizza decorator üst sınıfını oluşturuyoruz
class Decorator(Pizza):
   def __init__(self, component):
    self.component=component

   def get_cost(self):
        return str(self.component.get_cost() + Pizza.get_cost(self))

   def get_description(self):
        return self.component.get_description() + ' ' + Pizza.get_description(self)

#Decorator sosların her birini bir sınıf olarak tanımlıyoruz
class Zeytin(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 2
        self.description = "Zeytin Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class Mantar(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 3
        self.description = "Mantar Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class KeciPeyniri(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 3
        self.description = "Keçi Peyniri Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class Et(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 5
        self.description = "Et Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class Sogan(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 1
        self.description = "Soğan Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price


class Misir(Decorator):
    def __init__(self, pizza):
        super().__init__(pizza)
        self.price = 2
        self.description = "Mısır Sosu"

    def get_description(self):
        return self.description

    def get_cost(self):
        return self.price
```

```python
#Main fonksiyonu
with open('Menu.txt') as f:
    Menu = f.readlines()
    for item in Menu:
       print(item.strip())

pizzaSecim=input("Pizzanızı seçiniz: ")
if pizzaSecim=="Classic":
        pizza= KlasikPizza()
elif pizzaSecim == "Margherita":
        pizza = MargaritaPizza()
elif pizzaSecim =="TurkPizza":
        pizza = TurkPizza()
elif pizzaSecim=="PlainPizza":
        pizza= DominosPizza()
else: print("Lütfen geçerli bir pizza seçimi yapınız.")

sosSecim=input("Sos seçiniz: ")
if sosSecim=="Olives":
    sos = Zeytin(pizza)
elif sosSecim == "Mushrooms":
      sos = Mantar(pizza)
elif sosSecim == "GoatCheese":
      sos = KeciPeyniri(pizza)
elif sosSecim == "Meat":
      sos = Et(pizza)
elif sosSecim == "Onion":
      sos = Sogan(pizza)
elif sosSecim == "Corn":
      sos = Misir(pizza)
else: print("Lütfen geçerli bir sos secimi yapiniz.")

#Sipariş bilgileri
total_cost = pizza.get_cost()+sos.get_cost()
print("Siparişiniz: "+pizza.get_description()+" ve "+sos.get_description())
print(f"Siparişinizin toplam tutarı: {total_cost}")

#Kullanıcı bilgileri
name = input("Adınız: ")
id_no = input("TC Kimlik numaranızı giriniz: ")
card_no = input("Kredi kartı numaranızı giriniz: ")
card_cvv = input("CVV giriniz: ")

with open('OrdersDB.csv','w', newline='') as f:
  writer = csv.writer(f)
  writer.writerow([pizza.get_description(),sos.get_description(), total_cost, name, id_no, card_no, card_cvv, datetime.now().strftime("%Y-%m-%d %H:%M:%S")])
```
